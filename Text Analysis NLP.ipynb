```python
# Import Stopwords text files
```

```python
Geo=open('StopWords_Geographic.txt','r')
s1=Geo.readlines()
Geo.close()

A=open('StopWords_Auditor.txt','r')
s2=A.readlines()
A.close()

C=open('StopWords_Currencies.txt','r')
s3=C.readlines()
C.close()
currencies = [i.split('|')[0].strip() for i in s3]

D=open('StopWords_DatesandNumbers.txt','r')
s4=D.readlines()
D.close()

Gen=open('StopWords_Generic.txt','r')
s5=Gen.readlines()
Geo.close()

GenL=open('StopWords_GenericLong.txt','r')
s6=GenL.readlines()
GenL.close()

Names=open('StopWords_Names.txt','r')
s7=Names.readlines()
Names.close()



all_stopwords=s1+s2+currencies+s4+s5+s6+s7
print(len(all_stopwords))
```

```python
all_stopwords
```

```python
stop_words=[]
for i in all_stopwords:
    if i[-1]=='\n':
        stop_words.append(i[:-1])
    else:
        stop_words.append(i)
print (len(stop_words))
```

```python
stop_words
```

```python
# Converted all stopwords to lowercase() and set()
stopwords_list=[i.lower() for i in set(stop_words)]
len(stopwords_list)
```

```python
# Converted all text to lowercase()
input2['text']=input2['text'].str.lower()
```

```python
input2
```

```python
# Data Cleaning 
```

```python
import string
string.punctuation
```

```python
def text_cleaning(text):
    text_punc_removed=[char for char in text if char not in string.punctuation]
    text_punc_removed_join=''.join(text_punc_removed)
    text_punc_removed_join_clean=[word for word in text_punc_removed_join.split() if word not in stopwords_list]
    final_text=' '.join(text_punc_removed_join_clean)
    return final_text
```

```python
input2['clean_text']=input2['text'].apply(text_cleaning)
input2
```

```python
#import Positive and Negative words 

P=open("positive-words.txt")
Pos=P.readlines()
P.close()

N=open("negative-words.txt")
Neg=N.readlines()
N.close()
```

```python
Pos
```

```python
Positive_words=[]
for i in Pos:
    if i[-1]=='\n':
        Positive_words.append(i[:-1])
    else:
        Positive_words.append(i)
Positive_words
```

```python
Neg
```

```python
Negative_words=[]
for i in Neg:
    if i[-1]=='\n':
        Negative_words.append(i[:-1])
    else:
        Negative_words.append(i)
Negative_words
```

```python
len(Positive_words),len(Negative_words)
```

```python
# Calculate Positive and Negative Scores
```

```python
def scan_Positive(text):
    P_score=0
    for i in text.split():
        if i in Positive_words:
            P_score=P_score+1
    return P_score
```

```python
def scan_Negative(text):
    N_score=0
    for i in text.split():
        if i in Negative_words:
            N_score=N_score+1
    return N_score
```

```python
input2["POSITIVE SCORE"]=input2['clean_text'].apply(scan_Positive)
input2["NEGATIVE SCORE"]=input2['clean_text'].apply(scan_Negative)
input2["POLARITY SCORE"]=((input2["POSITIVE SCORE"] - input2["NEGATIVE SCORE"])/ ((input2["POSITIVE SCORE"] + input2["NEGATIVE SCORE"]) + 0.000001))
```

```python
#Subjectivity Score = (Positive Score + Negative Score)/ ((Total Words after cleaning) + 0.000001)
```

```python
input2['total_clean_text']=input2['clean_text'].apply(lambda txt: textstat.lexicon_count(txt))
```

```python
input2["SUBJECTIVIITY SCORE"]=((input2["POSITIVE SCORE"] + input2["NEGATIVE SCORE"])/ ((input2['total_clean_text']) + 0.000001))
```

```python
#Analysis of Readability
```

```python
input2['number of words']=input2['text'].apply(lambda txt: textstat.lexicon_count(txt))
input2['number of sentences']=input2['text'].apply(lambda txt: textstat.sentence_count(txt))
input2['Average Sentence Length'] = input2['number of words'] / input2['number of sentences']
```

```python
input2['COMPLEX WORD COUNT']=input2['text'].apply(lambda txt: textstat.difficult_words(txt))
input2['PERCENTAGE OF COMPLEX WORDS']=(input2['COMPLEX WORD COUNT']/input2['number of words'])*100
input2['FOG INDEX']=0.4 * (input2['Average Sentence Length'] + input2['PERCENTAGE OF COMPLEX WORDS'])
```

```python
#Average Number of Words Per Sentence
input2['AVG NUMBER OF WORDS PER SENTENCE']= input2['number of words'] / input2['number of sentences']
```

```python
#count the total cleaned words 
input2['WORD COUNT']=input2['clean_text'].apply(lambda txt: textstat.lexicon_count(txt))
```
